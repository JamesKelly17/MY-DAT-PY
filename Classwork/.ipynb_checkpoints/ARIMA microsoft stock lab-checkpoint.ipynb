### Predicting the Price of Microsoft Stock

This class will look to build an ARIMA(p, d, q) model from a history of Microsoft's stock price.

##### Step 1:  Load in the data set

```python
# Your Code Here
import numpy as np
import pandas as pd
%matplotlib inline

import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
plt.style.use('fivethirtyeight')

#Calculating Autocovariance and Autocorrelation in Statsmodels
from statsmodels.tsa.stattools import acf, acovf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

df = pd.read_csv('C:/Users/James.Kelly/DAT-06-24/class material/Unit 4/Class 18/data/msft.csv', index_col='Date', parse_dates=True)
df.tail()
```

##### Step 2:  Turn It Into A Stationary Time Series (You can do the same process that was used before)

```python
df.plot()
```

```python
autocorrelation_plot(df);
```

```python
df['Log_Price'] = np.log(df["Close"])
autocorrelation_plot(df['Log_Price'].diff()[1:]);
```

##### Step 3:  Plot the new Time Series, as well as the autocorrelation_plot in pandas

```python
df['Log_Price'].diff().plot();
```

```python
autocorrelation_plot(df['Log_Price'].diff()[1:]);
```

##### Step 4:  Using statsmodels, grab the acf and pacf plots to get an idea of what your p and q terms will be

```python
#q
stationary_data = df['Log_Price'].diff()[1:]
x = plot_acf(stationary_data, lags=40)
```

```python
#p

y = plot_pacf(stationary_data, lags=40)
```

##### Step 5: Fit the Model

```python
import warnings
warnings.simplefilter('ignore')

from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(df['Log_Price'], order=(2, 1, 1))
results = model.fit()
```

```python
results.aic
```

##### Step 6:  Pull Its Summary

```python
print(results.summary())

#minimize AIC score by changing p,d,q scores
```
