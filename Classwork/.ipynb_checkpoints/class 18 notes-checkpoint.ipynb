```python
import numpy as np
import pandas as pd
%matplotlib inline

df = pd.read_csv('C:/Users/James.Kelly/DAT-06-24/class material/Unit 4/Class 18/data/births.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
df.plot(title="Daily Births in CA 1959");
```

```python
#take all values until 1 day before the end, and subtract by mean of births

X_t = df[:-1]['Births'] - df['Births'].mean()
X_t.head()
```

```python
#all days, expect day #1

X_t_one = df[1:]['Births'] - df['Births'].mean()
X_t_one.head()
```

```python
X_covar = X_t.values * X_t_one.values
X_covar
```

```python
X_covar = X_covar.sum() * (1 / len(df))
X_covar
```

```python
#Calculating Convariance With Lag of 10

X_t = df[:-10]['Births'] - df['Births'].mean()
X_t_one = df[10:]['Births'] - df['Births'].mean()
X_covar = X_t.values * X_t_one.values
X_covar = X_covar.sum() * (1 / len(df))
X_covar
```

```python
#cheat shete formula

Xi = df['Births']
N = np.size(Xi)
k = 10
Xs = np.average(Xi)

def autocovariance(Xi, N, k, Xs):
    autoCov = 0
    for i in np.arange(0, N-k):
        autoCov += ((Xi[i+k])-Xs)*(Xi[i]-Xs)
    return (1/(N))*autoCov

print("Autocovariance:", autocovariance(Xi, N, k, Xs))
```

```python
# this is the autocovariance at lag 0
df['Births'].var()
```

```python
# the autocovariance at lag 10 is the covariance at lag 10 divided by the covariance at lag 0
X_covar / df['Births'].var()
```

```python
#Calculating Autocovariance and Autocorrelation in Statsmodels
from statsmodels.tsa.stattools import acf, acovf
```

```python
# this is the autocorrelation function at each subsequent lag (each score is lag at 0,1,2....)
acf(df)
```

```python
# this is the autocovariance of our time series at different lags (each score is lag at 0,1,2....)
acovf(df)
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
plt.style.use('fivethirtyeight')

autocorrelation_plot(df);
```

## Auto-Regressive and Moving Average Processes

```python
# ARIMA (p,d,q)
#p = how many units of time you look back
#d= how many times you differenced
#q = how many noise terms you regress agaisnt 
```

```python
#Autoregession: predictive of what will happen next based off of old data (like linear regression)
#moving average: previously occuring levels of noise
```

```python
data.tail()
```

```python
# looking back 2 days, this is what our prediction would look like

prediction = 4.3 + (432*0.7) + (390*0.4)
prediction
```

```python
#Moving average: of previously occurin  levels of nosie
```

```python
#This is for the MA part (q)

#1 (possible two significant noise terms)

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

stationary_data = data['Passengers'].diff().diff(periods=12)[13:]
x = plot_acf(stationary_data, lags=40)
```

```python
#This is for the AR part (p)
# start with one unit of time looking back

y = plot_pacf(stationary_data, lags=40)
```

```python
import warnings
warnings.simplefilter('ignore')

from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(data['Passengers'], order=(1, 1, 1))
results = model.fit()
print(results.summary())
```

```python
# this makes a forecast for the next period in our time series
# first number is the forecast
# second number is the standard deviation of the forecast
# third number is the 95% confidence interval
results.forecast()
```

```python
results.forecast(steps=3)
```

```python
# this is the final score of our model -- the lower the better
results.aic
```

```python
#SARIMA MODEL
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

mod = SARIMAX(data['Passengers'], order=(1, 1, 1), seasonal_order=(1, 1, 0, 12))
results = mod.fit()
print(results.summary())
```

```python
x = results.plot_diagnostics(figsize=(12, 15))
```

```python
results.forecast(6)
```
