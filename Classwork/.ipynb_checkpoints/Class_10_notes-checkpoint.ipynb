```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.gofplots import qqplot

# This allows plots to appear directly in the notebook.
%matplotlib inline
plt.style.use('fivethirtyeight') 
```

```python
data = pd.read_csv('C:/Users/James.Kelly/DAT-06-24/class material/Unit 3/data/advertising.csv', index_col=0)
data.head()
```

```python
#z-score, and p-value

sm.stats.ztest(data['radio'], data['newspaper'])
```

```python
##most important for testing normality - your data vs. theortical normal data (shown in stdev's)
#radio is nomral between 1 stdev
from statsmodels.graphics.gofplots import qqplot
x = qqplot(data['radio'])
```

```python
x = qqplot(data['newspaper'])
```

```python
#infering normality objectively 
#Shapiro Wilk - most reliable for n<2000 and not multiple repeat values
#scipy.stats.shapiro 
#parametric test  - assumes normality
#shows data is not normal
#null hypothesis is that the data is normal

from scipy.stats import shapiro

shapiro(data['radio'])
```

```python
#Kolmogoorvov-Smirnov test
#non-parametric, normalized distribution (you have to standardize data)
#https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.kstest.html
#more conservative, this one fails to reject null 
#can also use it to find the best distribution for your data

from scipy.stats import kstest

norm_data = (data['radio'] - data['radio'].mean())/data['radio'].std()
norm_data.describe()
kstest(norm_data,cdf='norm')
```

```python
# Generate some data and randomize.



# Some people bought nothing, the others bought 
# with some distribution.
#https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html
data1 = [0] * 50
data1.extend(np.random.normal(14, 4, 150))
np.random.shuffle(data1)

# The second design hooked less people, 
# but those that were hooked bought more stuff.
data2 = [0] * 100
data2.extend(np.random.normal(20, 5, 100))
np.random.shuffle(data2)

# Make a DataFrame.
df = pd.DataFrame()
df["A"] = data1
df["B"] = data2

df.head() 
```

```python
df.describe()
```

```python
sm.stats.ztest(df['A'], df['B'])
```

```python
norm_data2 = (df['A'] - df['A'].mean())/df['A'].std()
norm_data2.describe()
kstest(norm_data2,cdf='norm')
```

```python
#data with zeros removed
shifted_data = df[df.A > 0]['A']
shifted_data = (shifted_data - shifted_data.mean()) / shifted_data.std()
test_stat, p_value = kstest(shifted_data, cdf='norm')
print(test_stat, p_value)
```

```python
norm_data3= (df['B'] - df['B'].mean())/df['B'].std()
norm_data3.describe()
kstest(norm_data3,cdf='norm')
```

```python
#t-test - no assumption of knowing population variance
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html 
#good for small samples
#good for regression results

from scipy.stats import ttest_ind
ttest_ind(df['A'], df['B'], equal_var=False)
```

```python
#ANOVA
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.f_oneway.html
#compariance variance between two groups, with variance within groups

from scipy.stats import f_oneway

f_oneway(data['radio'], data['TV'], data['newspaper'])
```
